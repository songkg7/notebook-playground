```python
import pandas as pd
```

```python
df = pd.read_csv('data/merge_test.csv')
df
```

```python
df1 = df.iloc[:4, :]
df2 = df.iloc[4:, :]

display(df1)
display(df2)
```

```python
total = pd.concat([df1, df2])
total
```

```python
df3 = df.iloc[:2, :4]
df4 = df.iloc[5:, 3:]

display(df3)
display(df4)
```

```python
ans = pd.concat([df3, df4], join='inner')
ans
```

```python
ans = pd.concat([df3, df4], join='outer').fillna(0)
ans
```

```python
df5 = df.set_index('Location').T.iloc[:7, :3]
df6 = df.set_index('Location').T.iloc[6:, 2:5]

display(df5)
display(df6)
```

```python
ans = pd.merge(df5, df6, on='Algeria', how='inner')
ans
```

```python
ans = pd.merge(df5, df6, on='Algeria', how='outer')
ans
```
